```python
#podemos trasponer las tablas
area = pd.Series({'California': 423967, 'Texas': 695662,
                  'New York': 141297, 'Florida': 170312,
                  'Illinois': 149995})
pop = pd.Series({'California': 38332521, 'Texas': 26448193,
                 'New York': 19651127, 'Florida': 19552860,
                 'Illinois': 12882135})
info_geo = pd.DataFrame({'area':area, 'pop':pop})

#así podemos agregar otra columna
info_geo['density'] = info_geo['pop'] / info_geo['area']
info_geo

info_geo.T #la "T traspone"

#Para agregar espacios vacíos en necesario usar el comando "np.nan"
```

```python
#para saber si tenemos datos vacíos

tabla = pd.Series([1, np.nan, 'hello', None])
print(tabla)

print(tabla.isnull())

#para obtener sólo los datos no vacíos

tabla[tabla.notnull()]
```

```python
file = "salarios_minimos.csv" #hacemos la variable que va a contener ese nombre o en este caso documento
os.path.isfile(file) #te dice si existe aún o no la variable que pides en el mismo directorio
df_list = pd.read_html(html, header = 1)
df = df_list[-1]
df.to_csv(file)
df= pd.read_csv(file)
```
